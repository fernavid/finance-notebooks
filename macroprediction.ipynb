```python
import Quandl
import pandas
```

<h3>US Nominal GDP and the S&P 500</h3>

```python
gdp = Quandl.get("FRED/GDP")
market = Quandl.get("YAHOO/INDEX_GSPC")
```

<h3>Year over Year GDP Growth, Shifted Back in Time (sorcery!)</h3>

```python
future_gdp = gdp['Value'].pct_change(periods=4).shift(periods=-1) #YoY growth, known 1 quarter ahead of time!
market_return = market['Adjusted Close'].pct_change()
df = pandas.DataFrame({"Future GDP":future_gdp, 
                       "Market": market_return, 
                       "Strategy": market_return}, index=market.index)
df = df.fillna(method='ffill').dropna()
```

<h3>Invest in the S&P 500 When the Future Quarter's Growth is Above it's 5 Year Average, out of the Market Otherwise</h3>

```python
df.ix[df['Future GDP'] < pandas.stats.moments.rolling_mean(df['Future GDP'], 252*5), 'Strategy'] = 0
highgdp = df['Strategy']
highgdp_el = highgdp.add(1).cumprod()
bench_el = df['Market'].add(1).cumprod()
pandas.DataFrame({"Perfect Economic Vision":highgdp_el, 
                  "S&P 500":bench_el}).plot(logy=True)
```
