<h1>IPython Parallel Trading Strategy Optimization Example</h1>
Shows how to use Parallel computing to speed up parameter selection in a simple trading model. We start with a small problem and then show how the same principle pays off even more with larger, more complex problems with more parameters and more stocks.
<hr>

Specify our matplotlib chart sizes

```python
from pylab import *
rcParams['figure.figsize'] = 12, 5
```

Start a load balanced view with IPython parallel

```python
from IPython.parallel import Client
rc = Client()
lview = rc.load_balanced_view()
lview.block = True #wait until every job finishes before returning the result
```

Pick a ticker to optimize on and save the data locally

```python
from pandas.io.data import DataReader
from datetime import datetime

target = "^GSPC"
goog = DataReader(target, "yahoo", datetime(1990,1,1))['Adj Close']
goog.to_csv("/home/data/"+target+".csv")
```

Specify an optimization function that takes the ticker and the smoothing period and returns the sharpe ratio of trading that strategy

```python
import pandas, math
from IPython.parallel import depend, require, dependent

@require('pandas', 'math')
def opt(c, target):
    raw = pandas.Series.from_csv("/home/data/"+target+".csv") #every thread needs this data
    smoothing = pandas.stats.moments.rolling_mean(raw, c)
    above_ma = raw[raw.shift(periods=2)>smoothing.shift(periods=2)] #prevent lookahead bias
    returns = raw.pct_change().reindex(above_ma.index).reindex(raw.index).fillna(value=0)
    return [c, math.sqrt(250) * returns.mean() / returns.std(), target]
```

Use the load balanced view to test every smoothing period distributed among your cluster. 
It will return a list of results in the form [smoothing_period, sharpe_ratio]

```python
r = lview.map(opt, range(2, 252), [target for x in range(2, 252)])
```

Put the results into a pandas DataFrame, using the smoothing period as the index

```python
df = pandas.DataFrame(r, columns=['Moving Average Period Smoothing', 'Strategy Sharpe Ratio', 'Ticker'])
df = df.set_index('Moving Average Period Smoothing')

goog_returns = goog.pct_change() #benchmark returns for comparison purposes

df['Benchmark Sharpe Ratio'] = pandas.Series(math.sqrt(250) * goog_returns.mean() / goog_returns.std())
df.plot(title="Moving Average Strategies for "+target)
```

Putting it all together. We choose a smoothing period from an attractive part of the search space graph above and calculate the strategy equity line associated with trading that strategy. This code is very similar to the function used to optimize the strategy.

```python
best_smoothing = 150
raw = pandas.Series.from_csv("/home/data/"+target+".csv")
smoothing = pandas.stats.moments.rolling_mean(raw, best_smoothing)
above_ma = raw[raw.shift(periods=2)>smoothing.shift(periods=2)] #prevent lookahead bias
returns = raw.pct_change().reindex(above_ma.index).add(1).reindex(raw.index).fillna(value=1)
strategy = returns.cumprod()
benchmark = raw.pct_change().add(1).cumprod() #rebase to the strategy initial value
plt.figure()
strategy.plot(label='Strategy')
benchmark.plot(label='Benchmark')
plt.legend()
```

So now lets do this same analysis on every stock in the current Dow Jones Industrial Average. Survivorship bias ahead! Expect every strategy to do better than you think. More interesting to look at the structure and distribution of the strategy performance.

```python
stock_table = pandas.io.html.read_html("http://finance.yahoo.com/q/cp?s=^DJI", 
                                       attrs = {'class':'yfnc_tableout1'},
                                       header=0)
dj30_stocks = [y for x, y in stock_table[0]['Symbol'].iteritems()]

for stock in dj30_stocks:
    DataReader(stock, "yahoo", datetime(1990,1,1))['Adj Close'].to_csv("/home/data/"+stock+".csv")
    
```

We will build every combination instead of using the same ticker each time. We will alter our opt function a bit to make it compatible with our permutations list.

```python
import itertools

permutations = list(itertools.product(range(2,500), dj30_stocks))

@require('pandas', 'math')
def opt(params):
    c = params[0]
    target = params[1]
    raw = pandas.Series.from_csv("/home/data/"+target+".csv") #every thread needs this data
    smoothing = pandas.stats.moments.rolling_mean(raw, c)
    above_ma = raw[raw.shift(periods=2)>smoothing.shift(periods=2)] #prevent lookahead bias
    returns = raw.pct_change().reindex(above_ma.index).reindex(raw.index).fillna(value=0)
    return [c, math.sqrt(250) * returns.mean() / returns.std(), target]

r = lview.map(opt, permutations)
```

Now we can groupby the smoothing period and see the average sharpe for each smoothing!

```python
df = pandas.DataFrame(r, columns=['Moving Average Period Smoothing', 'Average Strategy Sharpe Ratio', 'Ticker'])
#df = df.set_index('Moving Average Period Smoothing')
average_sharpe = df.groupby("Moving Average Period Smoothing").mean()
average_sharpe.plot(title="Moving Average Strategies for Dow Jones Stocks")
```

Cool! So the longer the moving average the better the sharpe ratio. But how much variation is there for every stock? Let us plot every stocks performance by smoothing and see the spread in outcomes.

```python
df.set_index("Moving Average Period Smoothing").plot(title="Sharpe Ratio of all the Stocks by Smoothing Period")
```
